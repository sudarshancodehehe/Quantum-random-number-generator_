```python
# Function: Von Neumann correction
def von_neumann_correction(bits):
    corrected_bits = []
    i = 0
    while i < len(bits) - 1:
        if bits[i] != bits[i + 1]:
            corrected_bits.append(bits[i])
            i += 2
        else:
            i += 1  # Ignore consecutive bits
    return corrected_bits
```

```python
# Import necessary modules
from math import log2
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
import matplotlib.pyplot as plt

# Define helper functions
def binary_to_decimal(binary_string):
    return int(binary_string, 2)

def Random_no_generator(n):
    qc = QuantumCircuit(n)
    for i in range(n):
        qc.h(int(i))
    sv = qi.Statevector(qc)
    counts = sv.sample_counts(1)
    for key, value in counts.items():
        return binary_to_decimal(key)

def calculate_entropy(bits):
    prob_0 = bits.count(0) / len(bits)
    prob_1 = bits.count(1) / len(bits)
    if prob_0 == 0 or prob_1 == 0:  # Avoid log2(0)
        return 0
    entropy = - (prob_0 * log2(prob_0) + prob_1 * log2(prob_1))
    return entropy

# Generate random numbers
li = []
for _ in range(10240):  # Generate 10,240 random numbers
    li.append(Random_no_generator(5))

# Convert numbers to binary and collect all bits
binary_bits = []
for number in li:
    binary_bits.extend([int(bit) for bit in bin(number)[2:].zfill(5)])  # Convert to 5-bit binary

# Calculate entropy
entropy = calculate_entropy(binary_bits)
print(f"Entropy of the QRNG output: {entropy}")

# Calculate counts of 0s and 1s
count_0 = binary_bits.count(0)
count_1 = binary_bits.count(1)

# Plot the distribution of 0s and 1s
plt.figure(figsize=(8, 6))
plt.bar(["0", "1"], [count_0, count_1], color=["blue", "orange"], alpha=0.7)
plt.title(f"Distribution of Binary Bits\nEntropy: {entropy:.4f}", fontsize=14)
plt.xlabel("Bit Value", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi

# Define a QRNG function
def Random_no_generator(n):
    qc = QuantumCircuit(n)
    for i in range(n):
        qc.h(i)
    sv = qi.Statevector(qc)
    counts = sv.sample_counts(1)
    for key, value in counts.items():
        return int(key, 2)

# Generate random artwork
def generate_artwork(num_shapes=150, canvas_size=(10, 10)):
    fig, ax = plt.subplots(figsize=canvas_size)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis("off")  # Hide axes

    for _ in range(num_shapes):
        # Generate random properties for the shape
        shape_type = Random_no_generator(1) % 3  # 0 = circle, 1 = rectangle, 2 = triangle
        x = Random_no_generator(8) % 100 / 100  # Random x position
        y = Random_no_generator(8) % 100 / 100  # Random y position
        size = Random_no_generator(5) % 100 / 250 + 0.01  # Random size with minimum value
        color = (
            Random_no_generator(5) % 256 / 255,
            Random_no_generator(5) % 256 / 255,
            Random_no_generator(5) % 256 / 255,
            Random_no_generator(5) % 100 / 100,  # Random alpha transparency
        )

        # Draw the shape
        if shape_type == 0:  # Circle
            circle = plt.Circle((x, y), size, color=color, alpha=0.8)
            ax.add_artist(circle)
        elif shape_type == 1:  # Rectangle
            rectangle = plt.Rectangle(
                (x - size / 2, y - size / 2), size, size, color=color, alpha=0.8
            )
            ax.add_artist(rectangle)
        elif shape_type == 2:  # Triangle
            triangle = plt.Polygon(
                [
                    [x, y],
                    [x + size, y],
                    [x + size / 2, y + size * np.sqrt(3) / 2],
                ],
                color=color,
                alpha=0.8,
            )
            ax.add_artist(triangle)

    return fig

# Generate multiple artworks
def generate_multiple_artworks(num_artworks=8, num_shapes=150, canvas_size=(10, 10)):
    artworks = []
    for i in range(num_artworks):
        fig = generate_artwork(num_shapes=num_shapes, canvas_size=canvas_size)
        artworks.append(fig)
        plt.savefig(f"artwork_{i + 1}.png", bbox_inches="tight", dpi=300)  # Save artwork
        plt.close(fig)
    print(f"{num_artworks} artworks generated and saved as PNG files.")

# Generate and save 8 unique colorful artworks
generate_multiple_artworks(num_artworks=8, num_shapes=250, canvas_size=(10, 10))
```

## Blum, Blum, Shub (BBS) Algorithm
The Blum, Blum, Shub (BBS) algorithm is a pseudorandom number generator (PRNG) that is based on number theory and the difficulty of factoring large integers. It generates sequences of numbers that are computationally unpredictable, making it suitable for cryptographic applications.

# Binary reprsentation of the random numbers

```python
import math

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

def bbs_generate(p, q, seed, num_bits):
    # Verify primes are congruent to 3 mod 4
    if not (is_prime(p) and is_prime(q) and p % 4 == 3 and q % 4 == 3):
        raise ValueError("p and q must be primes congruent to 3 mod 4.")

    # Compute modulus
    M = p * q
    if gcd(seed, M) != 1:
        raise ValueError("Seed must be coprime to M = p * q.")

    # Initialize variables
    x = seed
    bit_sequence = []

    # Generate bits
    for _ in range(num_bits):
        x = pow(x, 2, M)  # x = x^2 mod M
        bit_sequence.append(x % 2)  # Extract least significant bit (LSB)

    return bit_sequence

# Example: Generate 20 random bits
p = 499  # Large prime
q = 547  # Large prime
seed = 12345  # Random seed
num_bits = 20

random_bits = bbs_generate(p, q, seed, num_bits)
print("Generated bits:", random_bits)
```
